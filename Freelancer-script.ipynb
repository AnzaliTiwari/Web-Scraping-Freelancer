```python
#To extract user information
def get_user_information(link):
    user_info = {}
    user_info['user_url'] = link
    driver.get(link) 
    page_new = driver.page_source
    page_soup_new = BeautifulSoup(page_new, 'html.parser')
    time.sleep(10)
    entry={
        'Name': name(),
        'Tagline': tagline(),
        'Rate': r(),
        'Location': loc(),
        'Rating': ratn(),
        'job_completed': jc(),
        'on_time': ot(),
        'rehire_rate': rh(),
        'certification': certi(),
        'skills': sk(),
        'highest_qualification': degree(),
        'College' : college(),
        'Current_Company': current_company(),
        'Role': job_Role(),
        'Company_second': company_2(),
        'Role_2': Role_2(), 
        'Company_third': company_3(),
        'Role_3': Role_3()
        
        
        
    }   
    return entry
    
```

```python
#function to get profiles(***added the feature of searching profiles of only India & according to the skills***)

def get_profiles(category,pages=1):
    flag=True
    count=1
    print(type(count))
    url = "https://www.freelancer.in/freelancers/india/{}/{}?w=f&ngsw-bypass="
    #"https://www.freelancer.in/freelancers/india/html?w=f&ngsw-bypass="
    url_new = url.format(category,count)
    driver.get(url_new)
    time.sleep(2)
    page = driver.page_source
    page_soup = BeautifulSoup(page, 'html.parser')
    columns=['Name','Tagline','Rate','Location','Rating','job_completed','on_time','rehire_rate','certification','skills','highest_qualification','College','Current_Company','Role','Company_second','Role_2','Company_third','Role_3']

    with open('demo_data.csv','a',encoding="utf-8") as f: #file name
        w=DictWriter(f,fieldnames=columns)
        w.writeheader()
        f.close()
    
    while flag:
        for user in page_soup.findAll('a',{"class":"find-freelancer-username"}):
            user_info = {}
            link = ""
            link = 'https://www.freelancer.in'+user['href']
            #print(link)
            columns=['Name','Tagline','Rate','Location','Rating','job_completed','on_time','rehire_rate','certification','skills','highest_qualification','College','Current_Company','Role','Company_second','Role_2','Company_third','Role_3']
            entry = get_user_information(link)
            with open('demo_data.csv','a',encoding="utf-8") as f: # file name as above
                w=DictWriter(f,fieldnames=columns)
                w.writerow(entry)
                            
        next_page = page_soup.find("ul",{"class":"Pagination"}).findAll("li",{"class":"Pagination-item PageFreelancersDirectory-pagination-item"})[-1]
        if next_page.find("a",{"class":"btn Pagination-link"}) is not None:
            #print("page exist")
            count += 1
            if count==pages+1:
                break
            #print(count)
            url_new = url.format(category,count)
            #print(url_new)
            driver.get(url_new)
            time.sleep(2)
            page = driver.page_source
            page_soup = BeautifulSoup(page,'html.parser')
            
        else:
            flag = False
    #return 
        

```
